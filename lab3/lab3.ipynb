# Завдання №1
### Згенерувати набір даних за допомогою означеної функції, з використанням якнайбільшої кількості параметрів.
### Розмірність набору даних прийняти у вигляді (100000х(30-n)), де n - номер за списком.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=100000, n_features=19, noise=2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Завдання №2
### Здійснити моделювання згенерованого набору даних та візуалізацію результатів.

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
print("Coefficients:\n",regressor.coef_)
y_pred = regressor.predict(X_test)
```

```python
plt.figure(figsize=(60, 30))
plt.scatter(range(0, len(y_test)), y_pred, label='true y')
plt.scatter(range(0, len(y_test)), y_test, label='pridection y')
plt.show()
```

```python

plt.plot(range(0, len(y_test)), y_pred, label='true y')
plt.plot(range(0, len(y_test)), y_test, label='pridection y')
plt.show()
```

# Завдання №3
### Надати значення метрик якості, приклад прогнозування

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
c_d = regressor.score(X_test, y_test)
print("Coefficient of determination: ", c_d)
```
